```python
import pandas as pd
import json

# Load the CSV data into a DataFrame
csv_file_path = 'data\Student_Performance_Table.csv'
data_frame = pd.read_csv(csv_file_path)

# Convert the DataFrame to JSON
json_string = data_frame.to_json(orient='records')

# Convert the JSON string to a Python object
data = json.loads(json_string)

json_file_path = 'Student_Performance_Table.json'
# Save the formatted JSON to a file with an indentation of 4 spaces
with open(json_file_path, 'w') as json_file:
    json.dump(data, json_file, indent=4)

print(f"CSV data has been converted and saved to '{json_file_path}'")
```
